```python
#Import neccessary packages
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.ensemble import RandomForestClassifier
from keras.callbacks import ModelCheckpoint
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix,recall_score,precision_score,accuracy_score,f1_score,roc_curve,roc_auc_score,auc,classification_report
```

```python
#Read in the train dataset
train_data = pd.read_csv('./train.csv')
train_data.head()
```

```python

train_data.info()
```

```python
#check for null values
train_data.isnull().sum()
```

Replace the missing values in Employment.Type with the mode 

```python
train_data['Employment.Type'] = train_data['Employment.Type'].fillna(train_data['Employment.Type'].mode()[0])
train_data.isnull().sum()
```

```python
#Check for the unique values in the ID columns
id_columns = ['UniqueID','branch_id','supplier_id','manufacturer_id','Current_pincode_ID','State_ID','Employee_code_ID']
for col in id_columns:
    print(col + "  " + str(train_data[col].nunique()))
```

```python
#Visualize the class Distribution
train_data.shape #view the shape of the dataset
sns.countplot(train_data['loan_default']) #plots the frequency count of each class
print(train_data.groupby('loan_default')['UniqueID'].count()) #returns count of each class
print('\n')

# check class distribution in percentage
count_0 = len(train_data[train_data["loan_default"] == 0])
count_1 = len(train_data[train_data["loan_default"]  == 1])
percentage_count_0 = ((count_0)/(count_0+count_1)) * 100
percentage_count_1 = 100-percentage_count_0
print("{}{}{}{}{}".format("Percentage of non_defaulters class is ",percentage_count_0,"\n","Percentage of defaulters class is ",percentage_count_1))
print('\n')

```

```python
#Drop the ID columns
#The manufacturer_id and state_ID might be important categorical features so they should not be dropped
columns_To_drop = ['UniqueID','branch_id','supplier_id','Current_pincode_ID','Employee_code_ID']
train_data = train_data.drop(columns_To_drop,axis =1)
```

```python
#Now i will be doin some type conversion.
#Convert all the flag columns to categorical along side the manufacturer id and the state_ID columns               
columns_toCategorical = ['manufacturer_id','State_ID','Passport_flag','Driving_flag','VoterID_flag','PAN_flag',
                      'Aadhar_flag','MobileNo_Avl_Flag']
for col in columns_toCategorical:
    train_data[col] = train_data[col].astype(str)
#convert columns with data tto DateTime
columns_toDate = ['Date.of.Birth','DisbursalDate']
for col in columns_toDate:
    train_data[col] = pd.to_datetime(train_data[col])
train_data.info()
```

```python
#Obtain Age as at the age of loanee as at the time of disbursment and drop the birthdate and disbursement date columns
train_data['Age'] = (train_data['DisbursalDate']-train_data['Date.of.Birth'])
train_data['Age_in_years'] = abs(round(train_data['Age']/np.timedelta64(1,'Y')))
train_data['Age_in_years'] = train_data['Age_in_years'].astype(int)
train_data['Age_in_years'].head(20)

```

```python
train_data = train_data.drop(['DisbursalDate','Date.of.Birth','AVERAGE.ACCT.AGE','CREDIT.HISTORY.LENGTH','Age'],axis =1)
train_data.info()
```

```python
train_data.describe()
```

```python
#Check correlation in the numeric features
numeric_columns = train_data.select_dtypes(include=['Int64','float64','int32'])
#Use Pearson coefficient to visualize numeric_feauture correlation
f, ax = plt.subplots(figsize=(14,12))
plt.title('Pearson Correlation of Numeric Features')
sns.heatmap(numeric_columns.iloc[:,0:23].astype(float).corr(), linewidths=0.25, vmax=1.0, square=True, 
           cmap=sns.diverging_palette(220, 10, as_cmap=True), ax=ax, linecolor='black', annot=True);
```

From the Above,it is obvious that some of the numeric features are highly correlated,i will drop one of each correlated feature and leave the other

```python
correlated_feature = ['SEC.SANCTIONED.AMOUNT','SEC.DISBURSED.AMOUNT','PRI.NO.OF.ACCTS','asset_cost']
train_data = train_data.drop(correlated_feature,axis =1)
train_data.info()
```

Now that i have dropped the highly correlated features,i will like to visualize the distribution of some of the variables

```python
sns.distplot(train_data["Age_in_years"], hist=True, kde=True,
             color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 1})
```

```python
sns.distplot(train_data["disbursed_amount"], hist=True, kde=True,
             color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 1})
```

Next i want to visualize class separation by the categorical features

```python
categorical_features = train_data.select_dtypes(include=['object'])
train_data['dummy'] = np.ones(shape = train_data.shape[0])
for col in categorical_features:
    print(col)
    counts = train_data[['dummy', 'loan_default', col]].groupby(['loan_default', col], as_index = False).count()
    temp = counts[counts['loan_default'] == 0][[col, 'dummy']]
    _ = plt.figure(figsize = (10,4))
    plt.subplot(1, 2, 1)
    temp = counts[counts['loan_default'] == 0][[col, 'dummy']]
    plt.bar(temp[col], temp.dummy)
    plt.xticks(rotation=90)
    plt.title('Counts for ' + col + '\n non-loanDefaulters')
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    temp = counts[counts['loan_default'] == 1][[col, 'dummy']]
    plt.bar(temp[col], temp.dummy)
    plt.xticks(rotation=90)
    plt.title('Counts for ' + col + '\n loanDefaulters')
    plt.ylabel('count')
    plt.show()
```
